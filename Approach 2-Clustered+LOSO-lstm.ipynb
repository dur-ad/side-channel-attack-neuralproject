```python
# ARIA Activity Recognition - LSTM Pipeline (Using Preprocessed Data)
# LSTM training with preprocessed train/test split

import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
from pathlib import Path
import json
import time
from tqdm.auto import tqdm

# Deep Learning
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, callbacks
from tensorflow.keras.utils import to_categorical

# ML utilities
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report, f1_score, accuracy_score, confusion_matrix

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Experiment tracking
import wandb
wandb.init(project="aria-activity-recognition-lstm", mode="offline")
from wandb.integration.keras import WandbCallback

# Set random seeds for reproducibility
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

print(f"âœ… TensorFlow version: {tf.__version__}")
print(f"âœ… GPU Available: {len(tf.config.list_physical_devices('GPU')) > 0}")

# ============================================================================
# CONFIGURATION
# ============================================================================
CONFIG = {
    'window_size': 256,
    'stride': 128,
    'seed': SEED,
    
    # LSTM Architecture
    'lstm_units': [128, 64],
    'dropout_rate': 0.3,
    'recurrent_dropout': 0.2,
    'dense_units': 64,
    
    # Training
    'batch_size': 64,
    'epochs': 50,
    'learning_rate': 0.001,
    'patience': 10,
    
    # Data
    'sensor_cols': ['acc_x', 'acc_y', 'acc_z', 'gyro_x', 'gyro_y', 'gyro_z', 'mag_x', 'mag_y', 'mag_z'],
    'target_col': 'cluster_right',  # Using cluster_right as target
    
    # Paths
    'train_path': '/kaggle/input/train-test-downsampled/train_down.parquet',
    'test_path': '/kaggle/input/train-test-downsampled/test_down.parquet'
}

print("âœ… Configuration loaded")

# ============================================================================
# DATA LOADING (Preprocessed Files)
# ============================================================================
def load_preprocessed_data(train_path, test_path):
    """Load preprocessed train/test data"""
    print("\nğŸ“‚ Loading preprocessed data...")
    
    print(f"   Loading train from: {train_path}")
    train_df = pd.read_parquet(train_path)
    
    print(f"   Loading test from: {test_path}")
    test_df = pd.read_parquet(test_path)
    
    print(f"\nâœ… Data loaded:")
    print(f"   Train: {len(train_df):,} records")
    print(f"   Test:  {len(test_df):,} records")
    
    # Show data info
    print(f"\nğŸ“Š Train data columns: {list(train_df.columns)}")
    print(f"\nğŸ“Š Target variable distribution (train):")
    print(train_df[CONFIG['target_col']].value_counts().sort_index())
    
    print(f"\nğŸ“Š Target variable distribution (test):")
    print(test_df[CONFIG['target_col']].value_counts().sort_index())
    
    return train_df, test_df

# ============================================================================
# SEQUENCE CREATION
# ============================================================================
def create_lstm_sequences(df, window_size, stride, sensor_cols, target_col):
    """Create sequences for LSTM"""
    print(f"\nğŸ”„ Creating LSTM sequences (window={window_size}, stride={stride})...")
    
    X_list = []
    y_list = []
    metadata_list = []
    
    # Group by script_id and sequence_id
    grouped = df.groupby(['script_id', 'sequence_id'])
    
    for (script_id, sequence_id), group in tqdm(grouped, desc="Processing sequences"):
        # Sort by timestamp
        group = group.sort_values('timestamp').reset_index(drop=True)
        
        if len(group) < window_size:
            continue
        
        # Extract sensor data
        sensor_data = group[sensor_cols].values
        
        # Get the most common activity in this group
        activity = group[target_col].mode()[0]
        
        # Create sliding windows
        for start_idx in range(0, len(group) - window_size + 1, stride):
            window = sensor_data[start_idx:start_idx + window_size]
            
            X_list.append(window)
            y_list.append(activity)
            metadata_list.append({
                'script_id': script_id,
                'sequence_id': sequence_id,
                'start_idx': start_idx
            })
    
    X = np.array(X_list, dtype=np.float32)
    y = np.array(y_list)
    
    print(f"âœ… Created {len(X):,} sequences")
    print(f"   Shape: {X.shape} (samples, timesteps, features)")
    print(f"   Class distribution: {np.unique(y, return_counts=True)}")
    
    return X, y, metadata_list

# ============================================================================
# PREPARE DATA FOR TRAINING
# ============================================================================
def prepare_training_data(train_df, test_df, config):
    """Prepare sequences and split validation set"""
    print("\nâš™ï¸ Preparing training data...")
    
    # Create sequences for train and test
    X_train_full, y_train_full, train_metadata = create_lstm_sequences(
        train_df,
        config['window_size'],
        config['stride'],
        config['sensor_cols'],
        config['target_col']
    )
    
    X_test, y_test, test_metadata = create_lstm_sequences(
        test_df,
        config['window_size'],
        config['stride'],
        config['sensor_cols'],
        config['target_col']
    )
    
    # Split train into train/val (90/10)
    print("\n   Splitting train into train/val (90/10)...")
    from sklearn.model_selection import train_test_split
    
    X_train, X_val, y_train, y_val = train_test_split(
        X_train_full, y_train_full,
        test_size=0.1,
        random_state=config['seed'],
        stratify=y_train_full
    )
    
    print(f"\nâœ… Data split:")
    print(f"   Train: {X_train.shape[0]:,} sequences")
    print(f"   Val:   {X_val.shape[0]:,} sequences")
    print(f"   Test:  {X_test.shape[0]:,} sequences")
    
    # Encode labels
    print("\n   Encoding labels...")
    label_encoder = LabelEncoder()
    label_encoder.fit(y_train_full)
    
    y_train_encoded = label_encoder.transform(y_train)
    y_val_encoded = label_encoder.transform(y_val)
    y_test_encoded = label_encoder.transform(y_test)
    
    n_classes = len(label_encoder.classes_)
    print(f"   Classes ({n_classes}): {label_encoder.classes_}")
    
    # Normalize data
    print("\n   Normalizing data...")
    n_samples, n_timesteps, n_features = X_train.shape
    
    scaler = StandardScaler()
    X_train_reshaped = X_train.reshape(-1, n_features)
    scaler.fit(X_train_reshaped)
    
    X_train = scaler.transform(X_train.reshape(-1, n_features)).reshape(n_samples, n_timesteps, n_features)
    X_val = scaler.transform(X_val.reshape(-1, n_features)).reshape(X_val.shape[0], n_timesteps, n_features)
    X_test = scaler.transform(X_test.reshape(-1, n_features)).reshape(X_test.shape[0], n_timesteps, n_features)
    
    # Convert to categorical
    y_train_cat = to_categorical(y_train_encoded, n_classes)
    y_val_cat = to_categorical(y_val_encoded, n_classes)
    y_test_cat = to_categorical(y_test_encoded, n_classes)
    
    print(f"\nâœ… Data preparation complete")
    
    return {
        'X_train': X_train, 'y_train': y_train_encoded, 'y_train_cat': y_train_cat,
        'X_val': X_val, 'y_val': y_val_encoded, 'y_val_cat': y_val_cat,
        'X_test': X_test, 'y_test': y_test_encoded, 'y_test_cat': y_test_cat,
        'scaler': scaler,
        'label_encoder': label_encoder,
        'n_classes': n_classes
    }

# ============================================================================
# LSTM MODEL ARCHITECTURE
# ============================================================================
def build_lstm_model(input_shape, n_classes, config):
    """Build LSTM model"""
    print("\nğŸ—ï¸ Building LSTM model...")
    
    model = keras.Sequential([
        layers.Input(shape=input_shape),
        
        layers.LSTM(
            config['lstm_units'][0],
            return_sequences=True,
            dropout=config['dropout_rate'],
            recurrent_dropout=config['recurrent_dropout'],
            name='lstm_1'
        ),
        layers.BatchNormalization(),
        
        layers.LSTM(
            config['lstm_units'][1],
            return_sequences=False,
            dropout=config['dropout_rate'],
            recurrent_dropout=config['recurrent_dropout'],
            name='lstm_2'
        ),
        layers.BatchNormalization(),
        
        layers.Dense(config['dense_units'], activation='relu', name='dense_1'),
        layers.Dropout(config['dropout_rate']),
        
        layers.Dense(n_classes, activation='softmax', name='output')
    ])
    
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=config['learning_rate']),
        loss='categorical_crossentropy',
        metrics=['accuracy', keras.metrics.AUC(name='auc')]
    )
    
    print("âœ… Model architecture:")
    model.summary()
    
    return model

# ============================================================================
# TRAINING
# ============================================================================
def train_model(model, data, config, run_name=None, use_wandb=False):
    """Train model with callbacks"""
    print("\nğŸš€ Training model...")
    
    callback_list = []
    
    if use_wandb:
        wandb.init(
            project='aria-activity-recognition-lstm',
            name=run_name,
            config={**config, 'model_type': 'LSTM'},
            tags=['lstm', 'keras', 'activity-recognition'],
            mode="offline"
        )
        callback_list.append(WandbCallback(monitor='val_loss', save_model=False))
    
    callback_list.extend([
        callbacks.EarlyStopping(
            monitor='val_loss',
            patience=config['patience'],
            restore_best_weights=True,
            verbose=1
        ),
        callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=5,
            min_lr=1e-6,
            verbose=1
        ),
        callbacks.ModelCheckpoint(
            'best_lstm_model.keras',
            monitor='val_loss',
            save_best_only=True,
            verbose=0
        )
    ])
    
    # Calculate class weights
    class_counts = np.bincount(data['y_train'])
    class_weights = {i: len(data['y_train']) / (len(class_counts) * count) 
                     for i, count in enumerate(class_counts)}
    
    print(f"   Class weights: {class_weights}")
    
    start_time = time.time()
    history = model.fit(
        data['X_train'], data['y_train_cat'],
        validation_data=(data['X_val'], data['y_val_cat']),
        epochs=config['epochs'],
        batch_size=config['batch_size'],
        class_weight=class_weights,
        callbacks=callback_list,
        verbose=1
    )
    training_time = time.time() - start_time
    
    print(f"âœ… Training complete: {training_time/60:.1f} minutes")
    
    return history, training_time

# ============================================================================
# EVALUATION
# ============================================================================
def evaluate_model(model, data, label_encoder):
    """Evaluate model"""
    print("\nğŸ“Š Evaluating model...")
    
    y_pred_probs = model.predict(data['X_test'], verbose=0)
    y_pred = np.argmax(y_pred_probs, axis=1)
    y_true = data['y_test']
    
    test_acc = accuracy_score(y_true, y_pred)
    test_f1_macro = f1_score(y_true, y_pred, average='macro')
    test_f1_weighted = f1_score(y_true, y_pred, average='weighted')
    
    print(f"ğŸ¯ Test Accuracy: {test_acc:.4f}")
    print(f"ğŸ¯ Test F1-Macro: {test_f1_macro:.4f}")
    print(f"ğŸ¯ Test F1-Weighted: {test_f1_weighted:.4f}")
    
    print("\nğŸ“‹ Classification Report:")
    report = classification_report(
        y_true, y_pred,
        target_names=label_encoder.classes_.astype(str),
        digits=4
    )
    print(report)
    
    cm = confusion_matrix(y_true, y_pred)
    
    return {
        'accuracy': test_acc,
        'f1_macro': test_f1_macro,
        'f1_weighted': test_f1_weighted,
        'confusion_matrix': cm,
        'predictions': y_pred,
        'probabilities': y_pred_probs
    }

# ============================================================================
# SAVE ARTIFACTS
# ============================================================================
def save_artifacts(model, scaler, label_encoder, config, metrics, history, test_data):
    """Save all artifacts comprehensively"""
    print("\nğŸ’¾ Saving artifacts...")
    
    artifacts_dir = Path("lstm_artifacts")
    artifacts_dir.mkdir(exist_ok=True)
    
    import joblib
    
    # 1. Save model
    print("   Saving model...")
    model.save(artifacts_dir / "lstm_model.keras")
    
    # 2. Save preprocessors
    print("   Saving preprocessors...")
    joblib.dump(scaler, artifacts_dir / "scaler.pkl")
    joblib.dump(label_encoder, artifacts_dir / "label_encoder.pkl")
    
    # 3. Save training history
    print("   Saving training history...")
    history_df = pd.DataFrame(history.history)
    history_df['epoch'] = range(1, len(history_df) + 1)
    history_df.to_csv(artifacts_dir / "training_history.csv", index=False)
    
    history_serializable = {k: [float(v) for v in vals] for k, vals in history.history.items()}
    with open(artifacts_dir / "training_history.json", 'w') as f:
        json.dump({'history': history_serializable, 'params': history.params}, f, indent=2)
    
    # 4. Save classification report
    print("   Saving classification report...")
    y_true = test_data['y_test']
    y_pred = metrics['predictions']
    
    report_text = classification_report(y_true, y_pred, target_names=label_encoder.classes_.astype(str), digits=4)
    with open(artifacts_dir / "classification_report.txt", 'w') as f:
        f.write("LSTM Model - Classification Report\n")
        f.write("="*80 + "\n\n")
        f.write(report_text)
        f.write(f"\n\nOverall Metrics:\n")
        f.write(f"  Test Accuracy:     {metrics['accuracy']:.4f}\n")
        f.write(f"  Test F1-Macro:     {metrics['f1_macro']:.4f}\n")
        f.write(f"  Test F1-Weighted:  {metrics['f1_weighted']:.4f}\n")
    
    report_dict = classification_report(y_true, y_pred, target_names=label_encoder.classes_.astype(str), 
                                       output_dict=True, zero_division=0)
    with open(artifacts_dir / "classification_report.json", 'w') as f:
        json.dump(report_dict, f, indent=2)
    
    # 5. Save test predictions
    print("   Saving test predictions...")
    predictions_df = pd.DataFrame({
        'true_label_encoded': y_true,
        'true_label': label_encoder.inverse_transform(y_true),
        'predicted_label_encoded': y_pred,
        'predicted_label': label_encoder.inverse_transform(y_pred),
        'correct': y_true == y_pred
    })
    
    for idx, class_name in enumerate(label_encoder.classes_):
        predictions_df[f'prob_class_{class_name}'] = metrics['probabilities'][:, idx]
    
    predictions_df['max_probability'] = metrics['probabilities'].max(axis=1)
    predictions_df.to_csv(artifacts_dir / "test_predictions.csv", index=False)
    
    # Save confusion matrix
    cm_df = pd.DataFrame(metrics['confusion_matrix'], 
                         index=label_encoder.classes_, 
                         columns=label_encoder.classes_)
    cm_df.to_csv(artifacts_dir / "confusion_matrix.csv")
    
    # 6. Save metadata
    print("   Saving metadata...")
    metadata = {
        'model_type': 'LSTM',
        'test_accuracy': float(metrics['accuracy']),
        'test_f1_macro': float(metrics['f1_macro']),
        'test_f1_weighted': float(metrics['f1_weighted']),
        'config': config,
        'input_shape': list(model.input_shape[1:]),
        'n_classes': len(label_encoder.classes_),
        'classes': label_encoder.classes_.tolist(),
        'total_parameters': int(model.count_params()),
        'training_epochs': len(history.history['loss']),
        'best_epoch': int(np.argmin(history.history['val_loss']) + 1),
        'best_val_loss': float(min(history.history['val_loss'])),
        'best_val_accuracy': float(max(history.history['val_accuracy']))
    }
    with open(artifacts_dir / "metadata.json", 'w') as f:
        json.dump(metadata, f, indent=2)
    
    # 7. Create README
    print("   Creating README...")
    readme_content = f"""# LSTM Model Artifacts

## Model Information
- **Model Type**: LSTM (Long Short-Term Memory)
- **Total Parameters**: {model.count_params():,}
- **Input Shape**: {model.input_shape[1:]}
- **Number of Classes**: {len(label_encoder.classes_)}

## Performance Metrics
- **Test Accuracy**: {metrics['accuracy']:.4f}
- **Test F1-Macro**: {metrics['f1_macro']:.4f}
- **Test F1-Weighted**: {metrics['f1_weighted']:.4f}

## Training Information
- **Total Epochs**: {len(history.history['loss'])}
- **Best Epoch**: {np.argmin(history.history['val_loss']) + 1}
- **Best Validation Loss**: {min(history.history['val_loss']):.4f}
- **Best Validation Accuracy**: {max(history.history['val_accuracy']):.4f}

## Classes
{chr(10).join(f'- {i}: {cls}' for i, cls in enumerate(label_encoder.classes_))}

Generated: {pd.Timestamp.now().strftime('%Y-%m-%d %H:%M:%S')}
"""
    with open(artifacts_dir / "README.md", 'w') as f:
        f.write(readme_content)
    
    print(f"\nâœ… All artifacts saved to: {artifacts_dir}")
    
    if wandb.run is not None:
        wandb.save(str(artifacts_dir / "lstm_model.keras"))
        wandb.save(str(artifacts_dir / "metadata.json"))
    
    return artifacts_dir

# ============================================================================
# VISUALIZATION
# ============================================================================
def plot_training_history(history, save_dir):
    """Plot training history"""
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    
    axes[0].plot(history.history['accuracy'], label='Train', linewidth=2)
    axes[0].plot(history.history['val_accuracy'], label='Validation', linewidth=2)
    axes[0].set_title('Model Accuracy', fontsize=14, fontweight='bold')
    axes[0].set_xlabel('Epoch')
    axes[0].set_ylabel('Accuracy')
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)
    
    axes[1].plot(history.history['loss'], label='Train', linewidth=2)
    axes[1].plot(history.history['val_loss'], label='Validation', linewidth=2)
    axes[1].set_title('Model Loss', fontsize=14, fontweight='bold')
    axes[1].set_xlabel('Epoch')
    axes[1].set_ylabel('Loss')
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)
    
    if 'auc' in history.history:
        axes[2].plot(history.history['auc'], label='Train', linewidth=2)
        axes[2].plot(history.history['val_auc'], label='Validation', linewidth=2)
        axes[2].set_title('Model AUC', fontsize=14, fontweight='bold')
        axes[2].set_xlabel('Epoch')
        axes[2].set_ylabel('AUC')
        axes[2].legend()
        axes[2].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig(save_dir / 'training_history.png', dpi=300, bbox_inches='tight')
    if wandb.run:
        wandb.log({"training_history": wandb.Image(str(save_dir / 'training_history.png'))})
    plt.show()

def plot_confusion_matrix(cm, class_names, save_dir):
    """Plot confusion matrix"""
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names,
                cbar_kws={'label': 'Count'})
    plt.title('Confusion Matrix - LSTM Model', fontsize=14, fontweight='bold')
    plt.xlabel('Predicted Class', fontsize=12)
    plt.ylabel('True Class', fontsize=12)
    plt.tight_layout()
    plt.savefig(save_dir / 'confusion_matrix_lstm.png', dpi=300, bbox_inches='tight')
    if wandb.run:
        wandb.log({"confusion_matrix": wandb.Image(str(save_dir / 'confusion_matrix_lstm.png'))})
    plt.show()

# ============================================================================
# MAIN PIPELINE
# ============================================================================
def main():
    """Execute full LSTM training pipeline"""
    print("="*80)
    print("ARIA ACTIVITY RECOGNITION - LSTM PIPELINE (PREPROCESSED DATA)")
    print("="*80)
    
    # 1. Load preprocessed data
    train_df, test_df = load_preprocessed_data(
        CONFIG['train_path'],
        CONFIG['test_path']
    )
    
    # 2. Prepare data (create sequences, normalize, encode)
    data = prepare_training_data(train_df, test_df, CONFIG)
    
    # 3. Build model
    input_shape = (CONFIG['window_size'], len(CONFIG['sensor_cols']))
    model = build_lstm_model(input_shape, data['n_classes'], CONFIG)
    
    # 4. Train
    history, training_time = train_model(model, data, CONFIG)
    
    # 5. Evaluate
    metrics = evaluate_model(model, data, data['label_encoder'])
    
    # 6. Save artifacts
    artifacts_dir = save_artifacts(
        model, data['scaler'], data['label_encoder'], 
        CONFIG, metrics, history, data
    )
    
    # 7. Visualize
    print("\nğŸ“Š Creating visualizations...")
    plots_dir = artifacts_dir / "plots"
    plots_dir.mkdir(exist_ok=True)
    
    plot_training_history(history, plots_dir)
    plot_confusion_matrix(metrics['confusion_matrix'], data['label_encoder'].classes_, plots_dir)
    
    # Final summary
    print("\n" + "="*80)
    print("TRAINING COMPLETE!")
    print("="*80)
    print(f"ğŸ¯ Test Accuracy: {metrics['accuracy']:.4f}")
    print(f"ğŸ¯ Test F1-Macro: {metrics['f1_macro']:.4f}")
    print(f"â±ï¸  Training Time: {training_time/60:.1f} minutes")
    print(f"ğŸ’¾ Artifacts saved to: {artifacts_dir}")
    print("="*80)
    
    wandb.finish()

# ============================================================================
# RUN
# ============================================================================
if __name__ == "__main__":
    main()
```
